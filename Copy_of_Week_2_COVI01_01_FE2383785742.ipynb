```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import os

file_name = list(uploaded.keys())[0]  # Get the first (or only) file name
file_path = os.path.join("/content", file_name)  # Use file_name instead of an undefined variable

print("File path:", file_path)
```

```python
# Define file path as a string
file_path = "/content/Corrupted Student Picture.png"

# Load the image
image = cv2.imread(file_path)  # Reads in BGR format

# Check if the image is loaded correctly
if image is None:
    print("Error: Could not load image. Check the file path!")
else:
    print("Image loaded successfully!")
```

```python
# Display Image
plt.imshow(image)
```

```python
# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Convert BGR to GRAY
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
# Convert BGR to HSV
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(image)
```

```python
# Define contrast (alpha) and brightness (beta) values
alpha = 1.5  # Adjust contrast
beta = 50    # Adjust brightness

# Apply contrast and brightness correction
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display the image
plt.imshow(adjusted_image)


```

```python
# Define contrast (alpha) and brightness (beta) values
alpha = 1.5  # Adjust contrast
beta = 20   # Adjust brightness

# Apply contrast and brightness correction
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Display the image
plt.imshow(adjusted_image)
```

```python
# Display original and restored images
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(image_rgb)
axs[0].set_title("Corrupted Image")
axs[0].axis("off")

axs[1].imshow(adjusted_image)
axs[1].set_title(f"Restored Image (α={alpha}, β={beta})")
axs[1].axis("off")
```
